```python
set1={1,2,3,4,5}
set2={4,5,8,9,7}
print('union:',set1|set2)
print('intersection:',set1&set2)
print('set1-set2:',set1-set2)
print('symmetric difference:',set1^set2)
```

```python
# Creating a dictionary
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "occupations": ["Engineer", "Artist"]
}
print("Original dictionary:", my_dict)

# Accessing elements by key
print("Name:", my_dict["name"])
print("Age:", my_dict.get("age")) # Using .get() is safer as it returns None if key not found

my_dict["age"] = 31
print("Dictionary after modifying age:", my_dict)

my_dict["email"] = "alice@example.com"
print("Dictionary after adding email:", my_dict)

removed_city = my_dict.pop("city")
print(f"Removed city: {removed_city}, Dictionary after removing city:", my_dict)

print("\nIterating through keys:")
for key in my_dict:
    print(key)

print("\nIterating through values:")
for value in my_dict.values():
    print(value)

print("\nIterating through key-value pairs:")
for key, value in my_dict.items():
    print(f"{key}: {value}")
print("\nLength of the dictionary:", len(my_dict))
```

```python
my_tuple = (1, 2, 3, "apple", "banana")
print("Original tuple:", my_tuple)
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])
another_tuple = (4, 5, 6)
combined_tuple = my_tuple + another_tuple
print("Combined tuple:", combined_tuple)
print("Length of the tuple:", len(my_tuple))
print("Slice from index 1 to 3:", my_tuple[1:4])
print("'apple' in my_tuple:", "apple" in my_tuple)
a, b, c, d, e = my_tuple
print(f"Unpacked elements: {a}, {b}, {c}, {d}, {e}")
```

```python
my_list = [1, 2, 3, "apple", "banana"]
print("Original list:", my_list)
print("First element:", my_list[0])
print("Last element:", my_list[-1])
my_list[0] = 10
print("List after modifying first element:", my_list)
my_list.append("cherry")
print("List after appending:", my_list)
my_list.remove("apple")
print("List after removing 'apple':", my_list)
print("Length of the list:", len(my_list))
print("Slice from index 1 to 3:", my_list[1:4])
```
